````python
 import matplotlib.pyplot as plt

# # Example: class1_images and class2_images should each have 6 images
# # Each image should be a NumPy array (H, W, C) or (H, W)
# # You can adapt this to load your own images as needed

 def visualize_two_classes(class1_images, class2_images, class1_label="Class 1", class2_label="Class 2"):
     assert len(class1_images) == len(class2_images), "Both classes must have the same number of images"

     num_images = len(class1_images)
     fig, axes = plt.subplots(2, num_images, figsize=(3 * num_images, 6))

     for i in range(num_images):
         # Row 0: class 1
         axes[0, i].imshow(class1_images[i], cmap='gray')
         axes[0, i].axis('off')
         if i == 0:
             axes[0, i].set_title(class1_label, fontsize=14, loc='left')

         # Row 1: class 2
         axes[1, i].imshow(class2_images[i], cmap='gray')
         axes[1, i].axis('off')
         if i == 0:
             axes[1, i].set_title(class2_label, fontsize=14, loc='left')

     plt.tight_layout()
     plt.show()
````

````python
import matplotlib.pyplot as plt
import random

#config
num_img_per_class = 5
class_labels = ["day", "night"] #0 = day 1 = night

#separate images by class
class1_img = [img for img, label in IMAGE_LIST if label == class_labels[0]]
class2_img = [img for img, label in IMAGE_LIST if label == class_labels[1]]

#Randomly select images from each class
selected_class1 = random.sample(class1_img, num_img_per_class)
selected_class2 = random.sample(class2_img, num_img_per_class)

#creat a figure to display the image
fig, axes = plt.subplots(2, num_img_per_class, figsize=(20, 6))
fig.suptitle("day and night images", fontsize=20)

for i in range(num_img_per_class):
    axes[0, i].imshow(selected_class1[i][0])
    axes[0, i].axis('off')
    axes[0, i].set_title("dayimage")

    axes[1, i].imshow(selected_class1[i][0])
    axes[1, i].axis('off')
    axes[1, i].set_title("nightimage")

plt.show()
````
